```python
# Make a plot with Matplotlib
%pip install matplotlib
%pip install numpy

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 1000)
y1 = np.cos(x)
y2 = np.sin(x)

plt.plot(x, y1, 'r')
plt.plot(x, y2, 'g')
plt.savefig('my_figure_name.png')
plt.show()
plt.close()
```

```python
# For scatter plots
x = np.linspace(0, 10, 15)
y1 = np.random.random(15)
y2 = np.random.random(15)

plt.plot(x, y1, 'Xr')
plt.scatter(x, y2, marker = 'o', color = 'b')

plt.savefig('A_scatter_plot.png')
plt.show()
```

```python
# For scatter plots
x = np.linspace(0, 10, 15)
y1 = np.random.random(15)
y2 = np.random.random(15)

fig = plt.figure()  # make a blank figure
ax = plt.axes()  # add some axes

ax.plot(x, y1, 'Xr', label = 'apples')
ax.scatter(x, y2, marker = 'o', color='b', label='oranges')

ax.legend() # add legend (use the labels set)

ax.set_xlim([-5, 15]) # sets limits of y-axis
ax.set_ylim([-0.5, 1.5]) # sets limits for y-axis

ax.set_title('random fruit data')

plt.show()
```

### Pandas
1. Provide the ML tools in scikit-learn with training data, feature matrix.
2. Pyhton package used in dealing with data structures of this type.
3. What is a features matrix?
    1. A two-dimensional table with each row containing a single observation or data-point, and each column containing a value representing a particular feature of that data-point.
    2. E.g., table of data in a spreadsheet.
    3. Imported into python
    4. CSV or comma separated value format. Text files in which values in each row are separated by commas, and a new row is just a line break.
4. Although NumPy arrays are computationally fast, they can only contain one data type.
5. Pandas Package provides the DataFrame class.
    1. Allows for the construction of tabular data with potentially different data types in each column - a feature matrix; with better computational performance compared to python lists.

### Pandas Series
1. There are two types of structures in Pandas
    1. DataFrame: Rows and Columns. Each column can have a different data-type (numbers, text, dates, etc.)
    2. Series: This is like a single column of data or list of values. It is one dimensional.

```python
### Import DataFrames and Series
%pip install pandas
import pandas as pd

list1 = [1.3, 1.4, 0.8, 2.6]
list2 = ['daisy', 'daisy', 'clover', 'daffodil']
list3 = ['daisy', 'daisy', 'clover', 'daffodil']

df = pd.DataFrame(data={'size': list1, 'species1': list2, 'species2': list3})
print(df)
print()
print(df.dtypes)
```

```python
%pip install scikit-learn

import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()

iris_df = pd.DataFrame(iris.data)

iris_df.columns = iris.feature_names

print(iris_df.head())
```
